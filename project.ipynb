# Step 5: Data Visualization

1. Price Distribution

```python
sns.histplot(df["price"],bins=50,kde=True)
plt.title("Distribution of BMW Prices")
plt.show()
```

### Observation 1
Most BMW cars are priced in the mid-range, while very high-priced cars are rare.

2. Boxplot - Price by Fuel Type

```python
sns.boxplot(x=df["fuelType"] , y=df["price"] , palette="Set2")
plt.title("Price VS Fuel Type")
plt.show()
```

### Observation 2
Diesel and petrol cars show wide price variation, while hybrid/electric BMWs are fewer but generally priced higher on average.

3. CountPlot - Models

```python
sns.countplot(y=df["model"],order=df["model"].value_counts().index)
plt.title("Car Count by Model")
plt.show()
```

### Observation 3
A few BMW models dominate the dataset (like 3 Series, 1 Series), while other models appear much less frequently.

4. Line plot - Avg Price Over Year

```python
df.groupby("model")["price"].mean().plot(marker="o")
plt.title("Average Price by Year")
plt.show()
```

### Observation 4
The average price of BMW cars decreases for older models and increases for newer year models.

5. Scatter Plot â€“ Price vs Mileage

```python
plt.scatter(df["mileage"],df["price"],alpha=0.5)
plt.title("Price VS Mileage")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

### Observation 5
There is a negative relationship between price and mileage â€” cars with higher mileage usually cost less.

6. Heatmap â€“ Correlation

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Observation 6
Correlation heatmap shows that year and engine size are positively correlated with price, while mileage is negatively correlated with price.

7. Pie Chart â€“ Fuel Types

```python
df["fuelType"].value_counts().plot.pie(autopct="%1.1f%%")
plt.ylabel("")
plt.title("Fuel Type Distribution")
plt.show()
```

### Observation 7
Diesel cars form the largest share, petrol follows, while hybrid/electric cars make up only a small portion.

8. Bar Plot â€“ Transmission Types

```python
df["transmission"].value_counts().plot(kind="bar", color="orange")
plt.title("Transmission Distribution")
plt.show()
```

### Observation 8
Automatic cars are slightly more common than manual cars in the dataset.

9. Violin Plot â€“ Price by Transmission

```python
sns.violinplot(x="transmission", y="price", data=df)
plt.title("Price Distribution by Transmission")
plt.show()
```

### Observation 9
Automatic BMWs generally have higher average prices compared to manual ones.

10. KDE Plot â€“ MPG Distribution

```python
sns.kdeplot(df["mpg"], shade=True, color="green")
plt.title("MPG Distribution")
plt.show()
```

### Observation 10
Most BMW cars have moderate MPG values; very high fuel efficiency is rare in the dataset.

11. Strip Plot â€“ Price by Fuel Type

```python
sns.stripplot(x="fuelType", y="price", data=df, jitter=True)
plt.title("Price Spread by Fuel Type")
plt.show()
```

### Observation 11
Fuel type clearly impacts price distribution â€” hybrids are generally higher priced, while petrol and diesel show wider variation.

12. Pairplot â€“ Numerical Columns

```python
sns.pairplot(df[["price", "mileage", "mpg", "engineSize"]])
plt.show()
```

### Observation 12
Pairplot shows relationships between numerical columns: newer cars with lower mileage and higher engine size tend to cost more.

13. Strip Plot â€“ Price vs Engine Size

```python
sns.stripplot(x="engineSize", y="price", data=df, jitter=True)
plt.title("Price vs Engine Size")
plt.show()
```

### Observation 13
Cars with larger engine sizes generally have higher prices, while smaller engines are less costly.

14. Barplot â€“ Avg Price by Model (Top 10)

```python
top_models = df["model"].value_counts().head(10).index
sns.barplot(x="model", y="price", data=df[df["model"].isin(top_models)], estimator=np.mean)
plt.xticks(rotation=45)
plt.title("Average Price of Top 10 Models")
plt.show()
```

### Observation 14
Among top 10 BMW models, premium ones like X5 are priced higher, while 1 Series and 3 Series are more affordable.

15. Rug Plot â€“ Mileage

```python
sns.rugplot(df["mileage"], height=0.05)
plt.title("Mileage Rug Plot")
plt.show()
```

### Observation 15
Most cars have mileage concentrated in a mid-range, with fewer very high mileage cars.

16. Joint Plot â€“ Price vs Mileage

```python
sns.jointplot(x="mileage", y="price", data=df, kind="hex")
plt.show()
```

### Observation 16
Jointplot confirms strong negative correlation between price and mileage.

17. Hexbin â€“ Mileage vs MPG

```python
df.plot.hexbin(x="mileage", y="mpg", gridsize=30, cmap="Blues")
plt.title("Mileage vs MPG")
plt.show()
```

### Observation 17
Hexbin plot shows an inverse trend: cars with higher MPG tend to have lower mileage.

18. Boxen Plot â€“ Price by Year

```python
sns.boxenplot(x="year", y="price", data=df)
plt.title("Price Distribution by Year")
plt.show()
```

### Observation 18
Boxen plot shows newer year cars consistently priced higher, while older years show wider spread at lower prices.

19. Catplot â€“ Transmission vs Fuel Type

```python
sns.catplot(x="fuelType", hue="transmission", data=df, kind="count")
plt.title("Transmission vs Fuel Type")
plt.show()
```

### Observation 19
Catplot shows that both fuel type and transmission influence counts â€” diesel-automatic cars are most common.

20. Lmplot â€“ Engine Size vs Price

```python
sns.lmplot(x="engineSize", y="price", data=df, scatter_kws={"alpha":0.3})
plt.title("Engine Size vs Price (Regression)")
plt.show()
```

### Observation 20
Regression line indicates that cars with bigger engine size usually command higher prices.

# Final Conclusion (Graph)

1. **Diesel cars dominate** the dataset, though petrol is also significant.  
2. **Newer models cost more**, while older ones are cheaper due to age and mileage.  
3. **Automatic cars generally have higher average price** than manual cars.  
4. **Mileage, year, and engine size strongly affect the price** of BMW cars.
